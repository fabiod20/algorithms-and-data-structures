# Hash Table Analysis
This notebook shows an overall analysis of the performance of **hash tables**, data structures highly efficient for the implementation of dictionaries. A dictionary is a dynamic set supporting the following operations: *insert*, *search* and *delete* of an element.

A hash table consist of a set of elements, each of whom is given a key. The access to a single element is done by the key. In the following analysis it is assumed that the keys are distinct.

# 1. Direct Addressing Table
A hash table generalizes the concept of **array**. An array may be used to implement a table in which the key of each element corresponds to the position of the element in the array.

Thanks to **direct addressing**, all the operations on an array run in O(1) time. However, given that the array size $m$ is typically equal to the cardinality of the key set $\left|K\right|$, this solution turns out to be memory inefficient if the effective number of keys $n$ is less than $\left|K\right|$.

### Insert

#### Code
```java
public boolean insert(Element element) {
    if(table[element.getKey()] == null){
        table[element.getKey()] = element;
        n = n + 1;
        return true;
    } else {
        return false;
    }
}
```

#### Analysis
The analysis was conducted considering the following parameters:
- $m$ array size: $m = 10^6$
- $n$ effective number of elements: $n \in \left[0, m-10^3\right]$
- *n_step* increment of $n$: *n_step* $= 10^3$ 
- *op_iteration* number of times that each operation is run, for each $n$: *op_iteration* $= 10$ 
- $\left|K\right|$ cardinality of the key set: $\left|K\right| = m$
- $P(k=i)$ probability distribution of the keys: $P(k=i) = \frac{1}{\left|K\right|}  \quad \forall i \in K$

# 2. Hash Table
If the actual number of keys $n$ is smaller then the cardinality of keys set $\left|K\right|$, **hash tables** are more efficient compared to direct addressing tables, from a memory usage perspective.

Instead of using keys as indexes of the array, hash tables use the key of an element to compute his index, by applying an **hash function**: <br>
<center>$h : K \rightarrow \{0, 1, \dots, m-1\}$</center>

Two different keys may be linked to the same array's index. This event is known as **collision** and it can be handled with one of two different approaches: *chaining* and *open addressing*.

## 2.1. Chained Hash Table
The **Chaining** method expects to store in a linked list all the elements with the same hash value $h(k)$. The j-th list contains all the elements for which it is: $h(k) = j$. In this way the hash table is composed of an array that contains $m$ pointers to $m$ lists. 

Considering doubly linked lists, the execution time of insert and delete operations is $O(1)$. Regarding search operation, it is necessary to distinguish the worst case from the average case:
- In the worst case, search operation runs in $O(n)$ time.
- In the average case, the execution time depends on how the hash function assigns keys to different lists. Under certain conditions, the average execution time is $O(1)$.

### Insert
```java
public boolean insert(Element element) {
    int i = hash(element.getKey());
    table[i].addFirst(element); // O(1)
    n = n + 1;
    return true;
}
```

### Search
```java
public Element search(int key) {
    Element e = new Element(key);
    int k = hash(key);

    int index = table[k].indexOf(e); // O(n)

    if(index != -1) {
        return table[k].get(index); // O(n)
    } else
        return null;
}
```

### Delete
```java
public boolean delete(Element element) {
    int i = hash((int) element.getKey());
    boolean result;
    result = table[i].remove(element);  // O(1)

    if(result) {
        n = n - 1;
        return true;
    } else
        return false;
}
```

## Simple Uniform Hashing
A condition needed to have average execution time of search operation equals to $O(1)$ is **simple uniform hashing**. It means that all the elements have the same probability of being assigned to any of the $m$ lists of the table. In other words, the probability that an element with key $i$ is assigned to a list $j$ is:
<center>$P(h(k_{i}) = j) = \frac{1}{m} \quad \forall i \in K, \forall j \in [0, m-1]$</center>

Simple uniform hashing alone is not enough to ensure execution time $O(1)$ for the average case. Under these conditions the execution times depend on a parameter called **load factor**, defined as the ratio between the number of keys actually stored and the number of lists of the table:
<center>$\alpha = \frac{n}{m}$</center>

Clearly, if $n = O(m)$, it is $\alpha = O(1)$. Therefore, if the number of lists is at least proportional to the number of elements stored, search operation runs on average in fixed time.

### 2.1.1. Division Chained Hash Table
One possible solution to build an hash function is the **division method**:
<center>$h(k) = k \mod m \qquad \forall k \in K$</center>

## 2.2. Open Addressing Hash Table
The **open addressing** method expects to store all the elements inside the array. Collisions are handled by allowing each item with a certain key to be stored in different positions.

For this purpose, the hash function can be extend:
<center>$h : K \times \{0, 1, \dots, m-1\} \rightarrow \{0, 1, \dots, m-1\}$</center>

Each element is given a sequence of positions $<h(k,0), h(k,1), ..., h(k,m-1)>$, called **probing sequence**. To access an element you must go over the whole sequence. This process is called **probing**.

As with direct addressing, the maximum number of items that can be stored is equal to the array size $m$. To fill entirely the array, the probing sequence must be a permutation of values $0, 1, \dots, m-1$.

### Insert
```java
public boolean insert(Element element) {
    int i = 0;
    int j = 0;

    do {
        j = hash(element.getKey(), i);

        if(element.equals(table[j])) {  // garantisce che le chiavi siano distinte
            return false;
        }

        if(table[j] == null || DELETED.equals(table[j])) {
            table[j] = element;
            n = n + 1;
            return true;
        } else {
            i = i + 1;
        }
    } while(i < m);
    return false;
}
```

### Search
```java
public Element search(int key) {
    int i = 0;
    int j = 0;

    do {
        j = hash(key, i);

        if(table[j] != null && table[j].getKey() == key) {
            return table[j];
        } else {
            i = i + 1;
        }
    } while(i < m && table[j] != null);
    return null;
}
```

### Delete
```java
public boolean delete(Element element) {
    int i = 0;
    int j = 0;

    do {
        j = hash(element.getKey(), i);

        if(table[j] != null && table[j].equals(element)) {
            table[j] = DELETED;
            n = n - 1;
            return true;
        } else {
            i = i + 1;
        }
    } while(i < m && table[j] != null);
    return false;
}
```

### 2.2.1. Linear Probing Hash Table
A first example of open addressing is the **linear probing**, in which the hash function is: 
<center>$h(k, i) = (h_1(k) + i) \mod m \qquad \forall i \in [0, m-1]$</center>
dove $h_1(k)$  una funzione di hash ausiliaria. <br>
In quest'analisi  stato scelto:
<center>$h_1(k) = k \mod m \qquad \forall k \in K$</center>

The main limitations of linear probing are:
- It use only $m$ of the $m!$ possible permutation. The probing sequens of an item with key $k$ is: 
<center>$<h_1(k), h_1(k)+1, \dots, m-1, 0, 1, \dots, h_1(k)-1>$</center>
As you can see, the value h_1(k)$ define the whole sequence. Distinct sequences differs only for the starting point.
- It suffers from a problem kown as *primary thickening*. It may appear an high concentration of items in a certain range of the array.

### 2.2.2. Quadratic Probing Hash Table
Another example of open addressing is **quadratic probing**, which use the following hash function:
<center>$h(k, i) = (h_1(k) + c_1 i + c_2 i^2) \mod m \qquad \forall i \in [0, m-1]$</center>
where $h_1(k)$ is an auxiliary function.<br>
In this analysis we choose:
<center>$h_1(k) = k \mod m \qquad \forall k \in K$</center>
<center>$c_1 = 1 , \quad c_2 = 3$</center>

As in linear probing, quadratic probing uses only $m$ of the $m!$ possible permutation and the value of $h_1(k)$ defines the whole probing sequence.

Moreover, this solution suffers from a problem kown as *secondary thickening*, which is less dangerous compared to primary thickening of linear probing.

### 2.2.3. Double Hashing
The last example of open addressing is **double hashing**,  which use the following hash function:
<center>$h(k, i) = (h_1(k) + i \cdot h_2(k)) \mod m \qquad \forall i \in [0, m-1]$</center>
where $h_1(k)$ and $h_2(k)$ are auxiliary functions.<br>
In this analysis we choose:
<center>$h_1(k) = k \mod m \qquad \forall k \in K$</center>
<center>$h_2(k) = 1 + (k \mod m-1) \qquad \forall k \in K$</center>

Unlike the prevoius solutions, double hashing uses $m^2$ of the $m!$ possible permutation. Moreover, the key defines not only the starting point, but also the offset to use during the ispection of the sequence.

### 2.2.4 Collisions
Here it is reported a collisions analysis of the open addressing approaches described before.  

The analysis was conduct considering the following parameters:
- $m = 10^4$
- $n \in \left[1, m\right]$
- $\left|K\right| = 10^8$
- $P(k=i) = \frac{1}{\left|K\right|}  \quad \forall i \in K$

````python
df_linear_collision = pd.read_csv("HashTable\Open\Linear\InsertCollision.csv", sep = '\t')
df_linear_collision = df_linear_collision.assign(alpha = lambda x: x.n/x.m)

df_quadratic_collision = pd.read_csv("HashTable\Open\Quadratic\InsertCollision.csv", sep = '\t')
df_quadratic_collision = df_quadratic_collision.assign(alpha = lambda x: x.n/x.m)

df_double_collision = pd.read_csv("HashTable\Open\Double\InsertCollision.csv", sep = '\t')
df_double_collision = df_double_collision.assign(alpha = lambda x: x.n/x.m)
````

#### Collisions Analysis
Plotting the number of collisions depending on the load factor $\alpha$, it can be seen that collisions increase as $\alpha$ move closer to $1$.

````python
ax = df_linear_collision.plot.line(x = 'alpha', y = 'collisions', title = 'Linear Probing')
df_quadratic_collision.plot.line(x = 'alpha', y = 'collisions', ax = ax, title = 'Quadratic Probing')
df_double_collision.plot.line(x = 'alpha', y = 'collisions', ax = ax, title = 'Double Hashing')

ax.legend(['Linear Probing', 'Quadratic Probing', 'Double Hashing'])
plt.ylabel('collisions')
plt.show()
````
